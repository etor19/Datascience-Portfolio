replacing the missing values of the dataset with its nearest neighbour. The reason being that , the countries have been repetative of 15 times since each year, the life expectancy is recorded 
And the life expectancy per each year increases by a very small margin, but in 2015, there has been a lot of jumps. In other words, the differnce in life expectancy from 2014 to 2015 is very huge hence replacing missing values with the mean will not be a great move.
we can replace it with the median or mode instead or the nearest neighbours

```python
from sklearn.impute import KNNImputer
```

```python
cols_to_impute = ['Life expectancy', 'Adult Mortality', 'Alcohol', 'Hepatitis B', 'BMI', 'Polio', 'Total expenditure', 'Diphtheria', 'GDP', 'Population', 'thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources', 'Schooling']

for col in cols_to_impute:
    if col == 'Life expectancy':
        imputer = KNNImputer(n_neighbors=5)
        df[col] = imputer.fit_transform(df[[col]])
    elif col in ['Adult Mortality', 'Alcohol']:
        df[col].fillna(df[col].mean(), inplace=True)
    elif col in ['thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources', 'Schooling']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
plt.figure(figsize = (30,30))
correlation = df.corr()
sns.heatmap(correlation, annot =True, cbar = True)
plt.title('Correlation Plot', fontsize = 30)
```
